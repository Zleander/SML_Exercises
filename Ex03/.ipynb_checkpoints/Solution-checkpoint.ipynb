```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import minimize
```

## Exercise 3

```python
def get_data(n, seed=42):
    np.random.seed(seed)
    
    X = np.random.uniform(low=0, high=2, size=(n, 1))
    epsilon = np.random.normal(0, 2, (n, 1))
    y = 2 * np.sin(2 * X) + 0.1 * epsilon
    
    return X, y
```

```python
xs, ys = get_data(100)
xs.shape, ys.shape
```

### a)

```python
# Create a new figure with new default axis
fig, ax = plt.subplots(1, 1)

ax.scatter(xs, ys)
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### b)

```python
def ridge_regression(xs, ys, lam=1):
    n, D = xs.shape
    g = (xs.T @ xs) + (n * lam * np.identity(D))
    w = np.linalg.inv(g) @ xs.T @ ys
    return w


def lasso_regression(xs, ys, lam=1):
    n, D = xs.shape
    loss = lambda w: (1/n) * np.linalg.norm(ys - xs @ w) + lam * np.linalg.norm(w, ord=1)
    w = minimize(loss, x0=np.ones(shape=(D,1))).x
    return w
```

```python
print(ridge_regression(xs, ys))
print(lasso_regression(xs, ys))
```

### c)

$\lambda = \{0.1, 1, 10\}$

Ridge Regression:

```python
lambdas = [0.1, 1, 10]
mse_per_lambda = [None, None, None]
ys_pred_per_lambda = [None, None, None]

for i, lam in enumerate(lambdas):
    w = ridge_regression(xs, ys, lam=lam)
    ys_pred = w * (np.sort(xs))
    mse = ((ys_pred - ys) ** 2).mean()

    ys_pred_per_lambda[i] = ys_pred
    mse_per_lambda[i] = mse
    
print(mse_per_lambda)
```

```python
fig, axis = plt.subplots(1, 3, figsize=(15, 5))

for ax, ys_pred in zip(axis, ys_pred_per_lambda):
    plt.sca(ax)  # enable axis
    ax.plot(xs, ys_pred, color='gray')
    ax.scatter(xs, ys)
    ax.set_xlabel('X')
axis[0].set_ylabel('Y')
  
plt.show()
```

Lasso Regression:

```python
mse_per_lambda = [None, None, None]
ys_pred_per_lambda = [None, None, None]

for i, lam in enumerate(lambdas):
    w = lasso_regression(xs, ys, lam=lam)
    ys_pred = w * xs
    mse = ((ys_pred - ys) ** 2).mean()

    ys_pred_per_lambda[i] = ys_pred
    mse_per_lambda[i] = mse
    ys_pred_per_lambda[i] = ys_pred
    mse_per_lambda[i] = mse 
    
print(mse_per_lambda)


fig, axis = plt.subplots(1, 3, figsize=(15, 5))

for ax, ys_pred in zip(axis, ys_pred_per_lambda):
    plt.sca(ax)  # enable axis
    ax.plot(xs, ys_pred, color='gray')
    ax.scatter(xs, ys)
    ax.set_xlabel('X')
axis[0].set_ylabel('Y')
plt.show()
```

### d)

```python
datasets = [(np.c_[xs, xs**2], ys) for xs, ys in
            (get_data(100, seed=seed) for seed in range(0, 100, 2))]
ridge_ws = []
lasso_ws = []
linear_ws = []
for xs, ys in datasets:
  ridge_ws.append(ridge_regression(xs, ys))
  lasso_ws.append(lasso_regression(xs, ys))
  linear_ws.append(ridge_regression(xs, ys, lam=0))

titles=['Ridge', 'Lasso', 'Linear']
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i, ws in enumerate([ridge_ws, lasso_ws, linear_ws]):
  axs[i].scatter(np.abs(ws)[:,0], np.abs(ws)[:,1])
  axs[i].set_title(titles[i])
  axs[i].set_xscale('log')
  axs[i].set_yscale('log')
plt.show()
       
```

### e)

Ridge regression

```python
lambdas = [0.001, 0.01, 0.1, 1, 10]
xs, ys = get_data(100)
xs.shape, ys.shape

Omega = np.c_[np.ones(shape=xs.shape), xs, xs**2]
```

```python
fig, axis = plt.subplots(1, 1)

# Plot line of mse as a function of lambda
mse = []
preds = []
for i in lambdas:
    w = ridge_regression(Omega, ys, lam=i)
    ys_pred = Omega @ w
    mse.append(((ys - ys_pred) ** 2).mean())
    preds.append(ys_pred)
plt.plot(lambdas, mse, marker='x')
plt.xscale('log')
plt.show()

best_index = np.argmin(mse)
best_error = mse[best_index]
best_lambda = lambdas[best_index]

print(f'best lambda: {best_lambda}')
print(f'mse: {best_error}')
```

```python
best_pred = preds[best_index]

fig, axis = plt.subplots(1, 1)
sorter = np.argsort(xs, axis=0)[:,0]

axis.scatter(xs, ys)
axis.plot(xs[sorter], best_pred[sorter], color='red')
plt.show()
```

Linear regresion

```python
lin_w = ridge_regression(Omega, ys, lam=0)
lin_pred = Omega @ lin_w
mse = ((ys - lin_pred) ** 2).mean()
print(mse)
```

```python
fig, axis = plt.subplots(1, 1)

axis.scatter(xs, ys)
axis.plot(xs[sorter], lin_pred[sorter] , color='red')
plt.show()
```
